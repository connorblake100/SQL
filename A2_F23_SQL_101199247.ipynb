#### 1. (4 pts) Retrieve all Album IDs, titles, and artist names where the artist name begins with 'A' (e.g., 'Alice In Chains').

```python
DataerFrame = pd.read_sql("""
SELECT 
    AlbumId, Title, Name as ArtistName 
FROM 
    Album 
INNER JOIN 
    Artist ON Album.ArtistId = Artist.ArtistId
WHERE 
    Artist.Name LIKE 'A%'
""",conn)

DataerFrame
```

#### 2. (6 pts) Write an SQL query that returns all the customer names and total amounts of their invoices (TotalAmount); Order the top 5 results highest to lowest by totalamount.

```python
df = pd.read_sql("""
SELECT 
    FirstName, LastName, SUM(Total) AS TotalAmount 
FROM 
    Customer
JOIN 
    Invoice 
    ON Customer.CustomerId = Invoice.CustomerId
GROUP BY 
    Customer.CustomerId, FirstName, LastName
ORDER BY 
    TotalAmount DESC 
LIMIT 
    5
""",conn)


df
```

#### 3. (5 pts) Retrieve the total number of invoices and the total revenue for each year.

```python
NEWdf = pd.read_sql("""
SELECT 
    strftime('%Y', InvoiceDate) AS Year,
    COUNT(InvoiceId) AS TotalInvoices,
    SUM(Total) AS TotalRevenue
FROM 
    Invoice
GROUP BY 
    Year
""",conn)
NEWdf
```

#### 4. (6 pts) Retrieve the total revenue for each genre, sorted in descending order of revenue.

```python
GenreDF = pd.read_sql("""
SELECT
    Total as GenreRevenue,
    Name as Genre
FROM 
    Invoice
JOIN
    Genre ON Invoice.InvoiceId = Genre.GenreId
ORDER BY
    GenreRevenue DESC
""",conn)

GenreDF 
    
```

#### 5. (7 pts) Write a query that returns the five top-selling tracks (track ID and track name, and Composer) and the quantity sold (TotalQuantitySold).

```python
TrackStarsTop5 = pd.read_sql("""
SELECT 
    Track.TrackId,
    Name AS TrackName,
    Composer,
    SUM(Quantity) AS TotalQuantitySold
FROM 
    Track
JOIN 
    InvoiceLine ON Track.TrackId = InvoiceLine.TrackId
GROUP BY 
    Track.TrackId
ORDER BY 
    TotalQuantitySold DESC
LIMIT 5
""",conn)
TrackStarsTop5
```

#### 6. (5 pts) Write a query that returns the average length of tracks in each genre; select genre id, name, and average length.

```python
AverageGenreTrackTime = pd.read_sql("""
SELECT 
    Genre.GenreId,
    Genre.Name as Genre,
    AVG(Milliseconds) as AverageLength
FROM
    Genre
JOIN 
    Track ON Genre.GenreId = Track.GenreId
GROUP BY
    Genre.GenreId
""",conn)
AverageGenreTrackTime
```

#### 7. (6 pts) Retrieve the IDs and names of artists with Rock and Metal tracks in the database.

```python
RockMetalArtists = pd.read_sql("""
SELECT
    Artist.ArtistId,
    Artist.Name as ArtistName,
    Genre.Name as RockMetal
FROM
    Artist
JOIN
    Genre ON Genre.GenreId = Artist.ArtistId
WHERE
    RockMetal LIKE 'Metal' or RockMetal LIKE 'Rock'
""",conn)
RockMetalArtists
```

#### 8. (9 pts) Write a query that returns the customer names who purchased tracks from the 'Rock' and 'Metal' genres.

```python
RockMetalCustomers = pd.read_sql("""
SELECT
    Customer.CustomerId,
    FirstName,
    LastName,
    Genre.Name as RockMetal
FROM 
    Customer
JOIN
    Genre ON Genre.GenreId = Customer.CustomerId
WHERE
    RockMetal LIKE 'Rock' or RockMetal = 'Metal'
""",conn)
RockMetalCustomers
```

#### 9. (6 pts) Retrieve the names of employees who have directly managed other employees (managers), along with the number of employees they manage.

```python
ManagerStats = pd.read_sql("""
SELECT 
    Manager.EmployeeId AS ManagerID,
    Manager.LastName AS ManagerLastName,
    Manager.Title AS ManagerTitle,
    COUNT(Employee.EmployeeId) AS NumberOfEmployeesManaged
FROM
    Employee AS Manager
JOIN
    Employee ON Manager.EmployeeId = Employee.ReportsTo
WHERE
    Manager.Title LIKE '%Manage%'
GROUP BY
    Manager.EmployeeId, Manager.LastName, Manager.Title
""",conn)
ManagerStats
```

#### 10. (6 pts) Write an SQL query retrieves the names of the top 5 playlists with the highest total invoice line amounts, where the invoice line amount is calculated by multiplying the unit price and quantity for each track in the playlist.

```python
Top5playlist = pd.read_sql("""
SELECT
    Playlist.Name AS PlaylistName,
    SUM(InvoiceLine.UnitPrice * InvoiceLine.Quantity) AS TotalInvoiceAmount
FROM
    Playlist
JOIN
    PlaylistTrack ON Playlist.PlaylistId = PlaylistTrack.PlaylistId
JOIN
    Track ON PlaylistTrack.TrackId = Track.TrackId
JOIN
    InvoiceLine ON Track.TrackId = InvoiceLine.TrackId
GROUP BY
    Playlist.Name
ORDER BY
    TotalInvoiceAmount DESC
LIMIT 5
""",conn)
Top5playlist
```
